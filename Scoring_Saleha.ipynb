```python
df["Pays_a_risque"]=np.where(df["Pays"].isin([36,76,792,365,188,608,158,320,218,862]), 1, 0)
```

### Ajout variable IMF

```python
continents = pd.read_excel("continents.xlsx", index_col = "Code numérique")
```

```python
df["Continent"] = df['Pays'].map(continents["Continent"])
```

```python
df["Continent"] = np.where(df["Pays"]==891, 'Europe', df["Continent"])
df["Continent"] = np.where(df["Pays"]==530, 'Amérique du Nord', df["Continent"])
df["Continent"] = np.where((df["Pays"]==999)|(df["Pays"]==900), 'Unknown', df["Continent"])
```

```python
summa= df.groupby("Pays").sum()[["fraude"]]
summa.sort_values(by="fraude")
```

```python
summa['FD'] = summa.index.map(continents["FD"])
```

```python
df["FD"] = df['Pays'].map(continents["FD"])
```

```python
df["FD"] = np.where((df["FD"].isna()) & (df["Continent"]=='Europe'), 0.502679705619812, df["FD"])
df["FD"] = np.where((df["FD"].isna()) & ((df["Continent"]=='Asie')|(df["Continent"]=='Océanie')), 0.350481420755386, df["FD"])
df["FD"] = np.where((df["FD"].isna()) & (df["Continent"]=='Afrique'), 0.1582290828228, df["FD"])
df["FD"] = np.where((df["FD"].isna()) & ((df["Continent"]=='Amérique du Nord')|(df["Continent"]=='Amérique du Sud')), 0.324424803256989, df["FD"])
df["FD"] = np.where((df["FD"].isna()) & (df["Continent"]=='Unknown'), 0.319519817829132, df["FD"])
```

### Variable date et heure

```python
df["h_split"] = df["Heure"].str.split(':', expand=True)[0]
df["h_split"] = df["h_split"].astype(int)
```

```python
df.groupby("h_split").sum()[["fraude"]].sort_values(by="fraude",ascending=False)
```

Variable matin/aprem/soir

```python
df.loc[(df["h_split"]>=6) & (df["h_split"]<12), "Parties_Jour"] = 'Matinée'
df.loc[(df["h_split"]>=12) & (df["h_split"]<18), "Parties_Jour"] = 'Après-midi'
df.loc[(df["h_split"]>=18) | (df["h_split"]<6), "Parties_Jour"] = 'Nuit'
```

```python
#concaténer date et heure

df["DateHeure"]=df["Date"]+" "+df["Heure"]
```

```python
from datetime import datetime

fmt = "%d/%m/%Y %H:%M:%S"
DT_fmt = []

for line in df["DateHeure"]:
    nv_fmt = datetime.strptime(line,fmt)
    DT_fmt.append(nv_fmt)
```

```python
df["DT_fmt"]=DT_fmt
```

Extraction du jour de la semaine puis variable binaire pour weekday/weekend

```python
wkday=[]

for line in df["DT_fmt"]:
    dte=datetime.date(line)
    day=dte.weekday()
    wkday.append(day)
```

```python
df["wkday"]=wkday
```

```python
#attention dans wkday le 0 correspond à lundi 

df["weekday"]=np.where(df["wkday"].isin([0,1,2,3,4]), 1, 0)
```

```python
#Extraction mois

fmt = "%b"
Mois = []

for line in df["DT_fmt"]:
    nv_mois = datetime.strftime(line,fmt)
    Mois.append(nv_mois)
    
df["Mois"]=Mois
```

```python
df = pd.concat([df,pd.get_dummies(df[['Continent', 'Parties_Jour', 'Mois']])],axis = 1)
```

```python
from sklearn.model_selection import train_test_split
from collections import Counter
from sklearn.linear_model import LogisticRegression
from numpy import where
from matplotlib import pyplot

from imblearn.under_sampling import CondensedNearestNeighbour
from imblearn.under_sampling import TomekLinks
from imblearn.over_sampling import SMOTE
```

### Oversampling

```python
over_sampler = SMOTE(kind='borderline1')
X_res, y_res = over_sampler.fit_resample(X_train, y_train)
print(f"Training target statistics: {Counter(y_res)}")
print(f"Testing target statistics: {Counter(y_test)}")
```

# Sélection de variables

```python
from sklearn.ensemble import RandomForestClassifier
from skopt import BayesSearchCV
from skopt.space import Real, Integer,Categorical
from timeit import default_timer as timer
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import StandardScaler, normalize

scaler = StandardScaler() 
scaled_Xtrain = scaler.fit_transform(X_res)  
scaled_Xtest = scaler.fit_transform(X_test)
```

```python
param_dict_bayes_RF = {
    'n_estimators' : Integer(10,200),
    'criterion': Categorical(['gini', 'entropy']), 
    'max_depth': Integer(3,10),
    'bootstrap':Categorical([True,False])}

NB_ITER = 5

opt = BayesSearchCV(RandomForestClassifier(),param_dict_bayes_RF,n_iter=NB_ITER,cv=5,scoring = 'roc_auc')

start = timer()  
opt.fit(X_res, y_res)
delais = timer()- start
best_param_opti_bayes = opt.best_params_
best_score_opti_bayes = opt.best_score_
all_result_opti_bayes = opt.cv_results_
print(" #### RECHERCHE DES PARAMETRES OPTIMAUX POUR LE RANDOM FOREST #### ")
print(" \n Durée de recherche en seconde :")
print(delais)
print("\n Paramètres optimaux : ")
best_param_opti_bayes =opt.best_params_
print(best_param_opti_bayes)
best_score_opti_bayes = opt.best_score_
print("\n Résultats  : " + str(best_score_opti_bayes))
```

```python
clf = RandomForestClassifier(n_estimators=55, random_state=0, n_jobs=-1, max_depth=9, bootstrap=True,criterion='entropy')

clf.fit(X_res,np.ravel(y_res,order='C'))
pred = clf.predict(X_test)
accuracy_score(y_test, pred)
```

```python
#Calcul des métriques

fpr_train_RF, tpr_train_RF, thresholds_train_RF = roc_curve(y_res, clf.predict_proba(X_res)[:,1])
roc_auc_train_RF = auc(fpr_train_RF, tpr_train_RF)

fpr_test_RF, tpr_test_RF, thresholds_test_RF = roc_curve(y_test, clf.predict_proba(X_test)[:,1])
roc_auc_test_RF = auc(fpr_test_RF, tpr_test_RF)

plt.figure()
lw = 2
plt.plot(fpr_train_RF, tpr_train_RF, color='darkorange',
         lw=lw, label='Train -RF - ROC curve (area = %0.2f)' % roc_auc_train_RF)
plt.plot(fpr_test_RF, tpr_test_RF, color='darkblue',
         lw=lw, label='Test -RF - ROC curve (area = %0.2f)' % roc_auc_test_RF)

plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Comparaison courve ROC Train ')
plt.legend(loc="lower right")
plt.show()
```
