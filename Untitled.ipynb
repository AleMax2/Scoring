```python
df["Continent"] = np.where(df["Pays"]==891, 'Europe', df["Continent"])
df["Continent"] = np.where(df["Pays"]==530, 'Amérique du Nord', df["Continent"])
df["Continent"] = np.where((df["Pays"]==999)|(df["Pays"]==900), 'Unknown', df["Continent"])
```

```python
df["FD"] = df['Pays'].map(continents["FD"])
```

```python
df["FD"] = np.where((df["FD"].isna()) & (df["Continent"]=='Europe'), 0.502679705619812, df["FD"])
df["FD"] = np.where((df["FD"].isna()) & ((df["Continent"]=='Asie')|(df["Continent"]=='Océanie')), 0.350481420755386, df["FD"])
df["FD"] = np.where((df["FD"].isna()) & (df["Continent"]=='Afrique'), 0.1582290828228, df["FD"])
df["FD"] = np.where((df["FD"].isna()) & ((df["Continent"]=='Amérique du Nord')|(df["Continent"]=='Amérique du Sud')), 0.324424803256989, df["FD"])
df["FD"] = np.where((df["FD"].isna()) & (df["Continent"]=='Unknown'), 0.319519817829132, df["FD"])
```

```python
### Encodage heure et jour semaine
```

```python
df["h_split"] = df["Heure"].str.split(':', expand=True)[0]
df["h_split"] = df["h_split"].astype(int)
```

```python
df.groupby("h_split").sum()[["fraude"]].sort_values(by="fraude",ascending=False)
```

Variable matin/aprem/soir

```python
df.loc[(df["h_split"]>=6) & (df["h_split"]<12), "Parties_Jour"] = 'Matinée'
df.loc[(df["h_split"]>=12) & (df["h_split"]<18), "Parties_Jour"] = 'Après-midi'
df.loc[(df["h_split"]>=18) | (df["h_split"]<6), "Parties_Jour"] = 'Nuit'
```

```python
df.head()
```

```python
df = pd.concat([df,pd.get_dummies(df[['Continent', 'Parties_Jour', 'Month']])],axis = 1)
```

```python
df.head()
```

```python
#concaténer date et heure

df["DateHeure"]=df["Date"]+" "+df["Heure"]
```

```python
from datetime import datetime

fmt = "%d/%m/%Y %H:%M:%S"
DT_fmt = []

for line in df["DateHeure"]:
    nv_fmt = datetime.strptime(line,fmt)
    DT_fmt.append(nv_fmt)
```

```python
df["DT_fmt"]=DT_fmt
```

Extraction du jour de la semaine puis variable binaire pour weekday/weekend

```python
wkday=[]

for line in df["DT_fmt"]:
    dte=datetime.date(line)
    day=dte.weekday()
    wkday.append(day)
```

```python
df["wkday"]=wkday
```

```python
#attention dans wkday le 0 correspond à lundi 

df["weekday"]=np.where(df["wkday"].isin([0,1,2,3,4]), 1, 0)
```

```python
#Extraction mois

df["Month"] = df["DT_fmt"].dt.month
```
